```python
class BankAccount:
    def __init__(self, holder_name, balance=0):
        self.name = holder_name
        self.balance = balance
    
    def deposit(self, amount):
        self.balance += amount
        print(f'INR {amount} deposited. New balance is INR {self.balance}')
    
    def withdraw(self, amount):
        if amount > self.balance:
            print('Insufficient funds! Please try a lower figure.')
        else:
            self.balance -= amount
            print(f'INR {amount} withdrawn. New balance is INR {self.balance}')
    
    def transfer(self, other, amount):
        if amount > self.balance:
            print('Insufficient funds! Please try a lower figure.')
        else:
            self.balance -= amount
            other.balance += amount
            print(f'INR {amount} transferred to {other}. New balance is INR {self.balance}')
    




        
    

    
```
