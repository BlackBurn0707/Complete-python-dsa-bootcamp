```python
def binary_to_decimal(binary_str):
    lst = []
    for letter in binary_str:
        lst.append(int(letter))
    lst = lst[::-1]
    num = 0
    for i in range(0, len(lst), 1):
        num += (2**i)*lst[i]
    return num

binary_to_decimal('1101')

```
