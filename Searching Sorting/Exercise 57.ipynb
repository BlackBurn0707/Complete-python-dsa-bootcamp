```python
def insertion_sort(lst):
    if len(lst) > 0:
        s_lst = [lst[0]]
        for i in range(1, len(lst), 1):
            for j in range(0, len(s_lst), 1):
                if lst[i] > s_lst[-1]:
                    s_lst.append(lst[i])
                    break
                elif lst[i] < s_lst[j]:
                    s_lst.insert(j, lst[i])
                    break
        return s_lst
    else:
        return lst

insertion_sort([])
```
