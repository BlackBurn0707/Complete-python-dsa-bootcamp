```python
def selection_sort(lst):
    for j in range(0, len(lst)-1, 1):
        for i in range(j, len(lst), 1):
            minimum = lst[j]
            index = j
            if lst[i] < minimum:
                minimum = lst[i]
                index = i
            lst[j], lst[index] = lst[index], lst[j]
    return lst

selection_sort([64, 25, 12, 22, 11])
```
